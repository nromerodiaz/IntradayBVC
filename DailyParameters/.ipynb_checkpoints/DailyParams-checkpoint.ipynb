# Non-observable market quality dimensions

---

In this notebooks we present the methodology applied in calculating daily parameters associated with market quality. These parameters are also called *non-observable* since we need to do an *ex-post* calculation to obtain them.

*Price impact* is one of this non-observable parameters. Its objective is to quantify the effect that market activity has on an asset's price. There are several ways to measure price impact. In this notebook, we will follow Kyle 1980, who argues that an asset's traded volume has a linear impact on its price.

In this model, the change of a stock's market price due to buy orders is given by:

$$\Delta P_{i, t, d} = \beta_{i, d}^1 D_{i, t, d}V_{i, t, d} + \varepsilon_{i, t, d}$$

where $\Delta P_{i, t, d}$ is the change in price of an asset $i$, of the transaction $t$ in day $d$. 

$D_{i, t, d}$ is the direction of the transaction. i.e. if the transaction was initiated by a buyer or a seller:
 * $D=1$ - The transaction was initiated by a buyer
 * $D=-1$ - The transaction was initiated by a seller

$V_{i, t, d}$ is the volume of the order

## I. Calculating price impact
---

**1. Direction of the transaction**

Since our database contains *trade and quote* (TAQ) data, we don't know explicitly if a transaction is begun by a buyer or a seller. Despite this, we can make several assumptions. By comparing the price of TRADE operations with the market price of the asset, we can approximate the two following scenarios:

 1. If the price of the order is **greater** than the mid-price, the transaction was initiated by a *buyer* ($D=+1$)
 2. If the price of the order is **less** than the mid-price, the transaction was initiate by a *seller* ($D=-1$)


```python
# We find initiating parties
init_party = InitiatingParty(data)
```

```python
# Checking our results
init_party.head()
```

**2. Estimation by regression**

Once we have the initiating parties, we can perform a linear regression to estimate Kyle's price impact $\beta_{i, d}^1$:

$$\Delta P_{i, t, d} = \beta_{i, d}^1 D_{i, t, d}V_{i, t, d} + \varepsilon_{i, t, d}$$

Going term by term, we first obtain the change in price ($\Delta P_{i, t, d}$) and the order flow $D_{i, t, d}V_{i, t, d}$. For more details about these parameters check `PreprocessPipeline.ipynb`

```python
# We calculate the necessary parameters for the regression
impact_params = ImpactParameters(init_party)
```

```python
# We check the results
impact_params[["iniciado", "delta_p", "order_flow"]].head()
```

Now, we apply the `KyleImpactRegression()` function

```python
# We perform the regression
kyle_impact_params = KyleImpactRegression(impact_params)
```

```python
# Checking the results
kyle_impact_params.head()
```

```python
# Graphical representation
pd.DataFrame(kyle_impact_params["coef_regresion"]).plot(figsize=(10, 8), marker="o")
plt.title("Kyle impact parameters", fontsize=20)
plt.ylabel(r"$\beta_{i, d}^1$", fontsize=20)
plt.xlabel("Fecha", fontsize=20)
plt.show()
```
